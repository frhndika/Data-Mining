```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import pandas as pd
```

```python
data=pd.read_excel("AHC.xlsx")
print(data)
```

```python
x=np.asarray(data)
print(x)
```

```python
z = linkage(x, 'average')
print(z)
```

```python
plt.figure(figsize=(20,8))
plt.title('Dendrogram Hierarchical Klastering')
plt.xlabel('data')
plt.ylabel('jarak')
dendrogram(z)
plt.show()
```
